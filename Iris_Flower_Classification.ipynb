```python
def plot_incorrect_predict(df_prediction,x_features,y_features):
    fig, axs= plt.subplots(2,2,figsize=(12,12))
    axs= axs.flatten()
    sns.scatterplot(x =x_features,y =y_features,data= df_prediction,hue ='prediction_label',ax=axs[0])
    sns.scatterplot(x =x_features,y =y_features,data= df_prediction,hue ='target_names',ax=axs[1])
    sns.scatterplot(x =x_features,y =y_features,data= df_prediction,hue ='correct_prediction',ax=axs[2])
    axs[3].set_visible(False)
    
    plt.show()
```

```python
plot_incorrect_predict(df_prediction,"petal length (cm)","petal width (cm)")
```
